```python
import pytesseract
from PIL import Image
from google.colab import files
import cv2
import matplotlib.pyplot as plt
```

```python
def perform_ocr(image_path):
    try:
        # Read image
        img = cv2.imread(image_path)

        # Preprocessing: convert to grayscale + threshold
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

        # OCR extraction
        text = pytesseract.image_to_string(thresh)
        return text
    except Exception as e:
        return f"Error during OCR: {e}"
```

```python
print("üìÇ Upload an image file containing text")
uploaded = files.upload()

```

```python
if uploaded:
    uploaded_file_name = list(uploaded.keys())[0]
    print(f"‚úÖ Uploaded image: {uploaded_file_name}")

    # Display image
    img = cv2.imread(uploaded_file_name)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Uploaded Image")
    plt.axis("off")
    plt.show()

    # Run OCR
    extracted_text = perform_ocr(uploaded_file_name)
    print("\n-- Extracted Text --\n")
    print(extracted_text)
else:
    print("‚ö†Ô∏è No image file uploaded")

```
